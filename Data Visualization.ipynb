```python
#kernel density function represent each of tick marks with gaussian basis function
#kernel density estimation 
sns.rugplot(dset2)
#set max and min  limit on x-axis
x_min=dset2.min() - 2
x_max=dset2.max() + 2

#set x-axis to be 100 equally spaced points
x_axis=np.linspace(x_min,x_max,100)
# Gaussian basis functions are used to approximate univariate data, and the underlying density being estimated is Gaussian,
#the optimal choice for h (that is, the bandwidth that minimises the mean integrated squared error) is

bandwidth=((4*dset2.std()**5) / (3*len(dset2)))**0.2

kernel_list=[]
#Create kernel for each point and append it to kernel_list
for data_point in dset2:
    kernel = stats.norm(data_point, bandwidth).pdf(x_axis)
    kernel_list.append(kernel)

# scaling the plot

    kernel=kernel/kernel.max()
    kernel=kernel*0.4

    plt.plot(x_axis,kernel,color='grey',alpha=0.4)
    plt.ylim(0,1)
```

```python
sum_of_kernel=np.sum(kernel_list,axis=)
fig=plt.plot(x_axis,sum_of_kernel,color='black')
plt.yticks([])
sns.rugplot(dset2)
```

## combining plot style .distplot() -------> (histogram + rug )+kernel density estimation plot
#### distplot always give hist and kde but if you also want rug plot, so just set rug = True

```python
d_f=randn(100)
sns.set(style='darkgrid',color_codes=True)
sns.distplot(d_f,bins=20)
```

```python
sns.distplot(d_f,bins=20,rug=True,hist=True)
```

```python
sns.distplot(d_f,bins=20,
             kde_kws={'color':'red','label':'KDE PLOT'},
             rug_kws={'color':'black','label':'RUG PLOT'},
             hist_kws={'color':'skyblue','label':'HIST PLOT'},hist=True,rug=True)
```

```python
#also work for Series
 
ser=pd.Series(d_f)
sns.distplot(ser,bins=20)
```
