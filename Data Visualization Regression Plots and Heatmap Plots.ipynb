```python
import pandas as pd
import numpy as np
from numpy.random import randn
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt 
from scipy import stats
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot('total_bill','tip',tips)
```

```python
sns.lmplot('total_bill','tip',tips,
          scatter_kws={'color':'indianred','edgecolor':'red'},
          line_kws={'color':'blue','lw':3})
```

```python
sns.lmplot('total_bill','tip',tips,order=3,
          scatter_kws={'color':'indianred','edgecolor':'red'},
          line_kws={'color':'blue','lw':3})
```

```python
tips['tip_pect']=100*(tips['tip']/tips['total_bill'])
```

```python
tips.head()
```

```python
sns.lmplot('tip_pect','tip',tips)
```

```python
sns.lmplot('size','tip_pect',tips)
sns.set(style='darkgrid',color_codes=True)
```

```python
sns.lmplot('tip','tip_pect',tips,x_jitter=9)
```

```python
sns.lmplot('size','tip_pect',tips,x_estimator=np.mean)
```

```python
tips.head()
```

```python
sns.lmplot('tip','tip_pect',tips,hue='sex',markers=['x','o'])
```

```python
sns.lmplot('total_bill','tip_pect',tips,hue='day')
```

```python
sns.lmplot('total_bill','tip_pect',tips,hue='time')
```

```python
url='https://en.wikipedia.org/wiki/Local_regression'
url='https://towardsdatascience.com/5-types-of-regression-and-their-properties-c5e1fa12d55e'
sns.lmplot('total_bill','tip_pect',tips,lowess=True,line_kws={'color':'indianred'})
```

## heatmap and clustered matricies

```python
flight_df=sns.load_dataset('flights')
flight_df.head()
```

```python
newdf=flight_df.sum()
newdf
```

```python
flight_pvt=flight_df.pivot('month','year','passengers')
flight_pvt
```

```python
sns.heatmap(flight_pvt)
```

```python
sns.heatmap(flight_pvt, annot=True,fmt='d')
```

```python
sns.heatmap(flight_pvt,center=flight_pvt.loc['June',1953])
```
